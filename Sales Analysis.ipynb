```python
#importing library
import pandas as pd
import os
```

```python
# Merging 12 months of data
files = [file for file in os.listdir('./Sales_Data')]

all_months_data = pd.DataFrame()

for file in files:
    df = pd.read_csv("./Sales_Data/"+file)
    all_months_data = pd.concat([all_months_data, df])
all_months_data.to_csv("all_data.csv", index=False)
```

```python
# Reading in updated DF
all_data = pd.read_csv("all_data.csv")
all_data.head()
```

```python
#dropping all nan values
nan_df = all_data[all_data.isna().any(axis=1)]
nan_df.head()

all_data = all_data.dropna(how='all')
all_data.head()
```

```python
#eliminating all the or in the Order Date column
all_data = all_data[all_data['Order Date'].str[0:2]!= 'Or']
```

```python
#Adding a month column
all_data['Month'] = all_data['Order Date'].str[0:2]
all_data['Month'] = all_data['Month'].astype('int32')
all_data.head()
```

```python
all_data['Quantity Ordered'] = pd.to_numeric(all_data['Quantity Ordered'])
all_data['Price Each'] = pd.to_numeric(all_data['Price Each'])
all_data.head()
```

```python
#Calculating the best month for sales? How much was earned that month?
all_data['Sales'] = all_data['Quantity Ordered'] * all_data['Price Each']
all_data.head()
```

```python
#Adding a city column from Purchase Address
def get_city(address):
    return address.split(",")[1].strip(" ")

def get_state(address):
    return address.split(",")[2].split(" ")[1]

all_data['City'] = all_data['Purchase Address'].apply(lambda x: f"{get_city(x)}  ({get_state(x)})")
all_data.head()
```

```python
results = all_data.groupby('Month').sum()
results
```

```python
import matplotlib.pyplot as plt
```

```python
months = range(1,13)
plt.bar(months, results['Sales'])
plt.xticks(months)
plt.xlabel('Sales in USD ($)')
plt.ylabel('Month in numbers')
plt.show()
```

```python
#Calculating Which city sold the most product?
results = all_data.groupby('City').sum()
results
```

```python
months = range(1,13)
cities = [city for city, df in all_data.groupby('City')]
plt.bar(cities, results['Sales'])
plt.xticks(months, rotation='vertical')
plt.xlabel('cities name')
plt.ylabel('Sales in USD($)')
plt.show()
```

```python
#Calculating  What time should we display advertisements to maximize likelihood of customer's buying product?
all_data['Order Date'] = pd.to_datetime(all_data['Order Date'])
```

```python
all_data['Hour'] = all_data['Order Date'].dt.hour
all_data['Minute'] = all_data['Order Date'].dt.minute
all_data.head()
```

```python
hours = [hour for hour, df in all_data.groupby('Hour')]
plt.plot(hours, all_data.groupby(['Hour']).count())
plt.xticks(hours)
plt.xlabel('Hour')
plt.ylabel('Number of orders')
plt.grid()
plt.show()
```

```python
#Calculating What products are most often sold together?
df = all_data[all_data['Order ID'].duplicated(keep=False)]
df['Grouped'] = df.groupby('Order ID')['Product'].transform(lambda x: ',' .join(x))
df = df[['Order ID', 'Grouped']].drop_duplicates()
df.head(10)
```

```python
from itertools import combinations
from collections import Counter

count = Counter()
 
for row in df['Grouped']:
    row_list = row.split(',')
    count.update(Counter(combinations(row_list,5)))

for key, value in count.most_common(10):
    print(key, value)
```

```python
#Calculating What product sold the most? Why was it sold the most?
product_groupby = all_data.groupby('Product')
quantity_ordered = product_groupby.sum()['Quantity Ordered']
products = [Product for Product, df in product_groupby]
plt.bar(products, quantity_ordered)
plt.xlabel('Quantity Ordered')
plt.ylabel('Products')
plt.xticks(products, rotation = 'vertical', size=9)
plt.show()
```

```python
prices = all_data.groupby('Product').mean()['Price Each']
print(prices)
```
